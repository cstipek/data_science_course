```python
#Setting the variables
batchSize=100
nEpochs=50
```

```python
#Setting the model
model = tf.keras.Sequential()

#Need to flatten image input
layer_0 = tf.keras.layers.Flatten(input_shape=np.shape(trImages[0]), name='Images')
model.add(layer_0)
```

```python
#Layer1 is the output layer
numUnits_L1=10
layer_1 = tf.keras.layers.Dense(units=numUnits_L1, activation=tf.nn.softmax, use_bias=True, name='Logistic')
model.add(layer_1)
```

```python
#Compile the model
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#Train the model
model_1_train=model.fit(x=trImages, y=trLabels, batch_size=batchSize, epochs=nEpochs,
                     shuffle=False, validation_split=0.0)
```

```python
plt.figure(1)
plt.plot(model_1_train.history['accuracy'], 'b-o')
plt.grid(True)
plt.xlabel('Epoch #')
plt.ylabel('Accuracy')

plt.figure(2)
plt.plot(model_1_train.history['loss'], 'r-s')
plt.grid(True)
plt.xlabel('Epoch #')
plt.ylabel('Loss')

plt.show()
```

```python
# Running model on test data now
model_1_test=model.fit(x=tImages, y=tLabels, batch_size=batchSize, epochs=nEpochs,
                     shuffle=False, validation_split=0.0)
```

```python
plt.figure(1)
plt.plot(model_1_test.history['accuracy'], 'b-o')
plt.grid(True)
plt.xlabel('Epoch #')
plt.ylabel('Accuracy')

plt.figure(2)
plt.plot(model_1_test.history['loss'], 'r-s')
plt.grid(True)
plt.xlabel('Epoch #')
plt.ylabel('Loss')

plt.show()
```

# Estimator Model Here

```python
import os
import time
import tensorflow as tf
import numpy as np
```

```python
#Predicting 10 categories
LABEL_DIMENSIONS = 10

(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
TRAINING_SIZE = len(train_images)
TEST_SIZE = len(test_images)
```

```python
train_images = np.asarray(train_images, dtype=np.float32) / 255
```

```python
#Convert the train and add channels
train_images = train_images.reshape((TRAINING_SIZE, 28, 28, 1))
test_images = np.asarray(test_images, dtype=np.float32) / 255
```

```python
#Convert the train images and add channels
test_images = test_images.reshape((TEST_SIZE, 28, 28, 1))
train_labels = tf.keras.utils.to_categorical(train_labels, LABEL_DIMENSIONS)
test_labels = tf.keras.utils.to_categorical(test_labels, LABEL_DIMENSIONS)
```

```python
#Cast the labels to float
train_labels = train_labels.astype(np.float32)
test_labels = test_labels.astype(np.float32)
print(train_images.shape)
print(train_labels.shape)
print(test_labels.shape)
```

```python
#Building a CNN model
inputs = tf.keras.Input(shape=(28,28,1))
x = tf.keras.layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu')(inputs)
x = tf.keras.layers.MaxPooling2D(pool_size=(2,2), strides=2)(x)
x = tf.keras.layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu')(inputs)
x = tf.keras.layers.MaxPooling2D(pool_size=(2,2), strides=2)(x)
x = tf.keras.layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu')(inputs)
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(64, activation='relu')(x)
predictions = tf.keras.layers.Dense(LABEL_DIMENSIONS, activation='softmax')(x)
model = tf.keras.Model(inputs=inputs, outputs=predictions)
model.summary()
```

```python
#Compile the model now
optimizer = tf.keras.optimizers.SGD()
model.compile(loss='categorical_crossentropy', optimizer = optimizer, metrics=['accuracy'])
```

```python
#Define a strategy
strategy = None
config = tf.estimator.RunConfig(train_distribute=strategy)
```

```python
#Convert model into a convenient estimator
estimator = tf.keras.estimator.model_to_estimator(model, config=config)
```

```python
#Define input functions for training and for testing
def input_fn(images, labels, epochs, batch_size):
    #convert the inputs to a dataset
    dataset=tf.data.Dataset.from_tensor_slices((images, labels))
    #Shuffle, repeat, and batch
    SHUFFLE_SIZE=5000
    dataset=dataset.shuffle(SHUFFLE_SIZE).repeat(epochs).batch(batch_size)
    dataset=dataset.prefetch(None)
    
    return dataset
```

```python
###PLEASE NOTE THIS CELL WILL TAKE AT LEAST 10-15 MINUTES###

BATCH_SIZE = 3
EPOCHS=3
estimator_train_result = estimator.train(input_fn = lambda:input_fn(train_images,
                                        train_labels, epochs=EPOCHS,
                                        batch_size=BATCH_SIZE))

print(estimator_train_result)
```

```python
#Evaluate
estimator_train_result.evaluate(lambda: input_fn(test_images,
                                   test_labels,
                                   epochs=1,
                                   batch_size=BATCH_SIZE))
```
